```python
import logreg
import xlrd
import numpy as np
from sklearn import linear_model

data = xlrd.open_workbook('melon.xlsx')
table = data.sheet_by_index(0)
dataFile = []
for rowNum in range(table.nrows):
    dataFile.append(table.row_values(rowNum))
```

```python
w = logreg.logreg(dataFile, 0.01)
print(w)
```

```python
logreg.logplot(w, dataFile)
```

```python
y_hat1 = logreg.predict(w, dataFile)
print(y_hat1)

data_len = len(dataFile)
par_len = len(dataFile[0])
y = []
for i in range(data_len):
    y.append(dataFile[i][par_len - 1])
    dataFile[i].pop()
y = np.trunc(y).astype(int).tolist()
print(y)

X = np.array(dataFile)
y = np.array(y)
classifier = linear_model.LogisticRegression(solver='liblinear',C=100)
classifier.fit(X, y)
y_hat2 = classifier.predict(X)
print(y_hat2)
```
